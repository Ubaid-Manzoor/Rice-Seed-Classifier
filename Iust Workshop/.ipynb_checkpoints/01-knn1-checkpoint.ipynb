## Classifiers using sklearn

__sklearn__ is a popular python based Machine learning library. It has a common format where we create a model object like `SVM`, `Decision Tree`, `GMM` etc and then fit the data over the model using a `fit()` function. For predictions as well we have uniform methods like `predict()` and `predict_proba()` across the models.
Many deep learning libraries are inspired from the design of __sklearn__

## Common Tasks

A particular ML task encompasses a few steps:
1. Import The Dependencies
2. Load the data
3. Split it into train, test and validation sets
4. An optional plotting to visualize the data.
5. Create the ML model object like `SVM`, `DecisionTree` etc by calling the constructor
6. Fitting the data by calling the `fit()` function on the training data
7. Evaluate the model on the validation data.
8. Repeat `6` and `7` till a better result is achieved.
9. Evaluate on the test data and report the results.

Below we will do all these tasks on different classifiers.

## Step 1: Import the dependencies
The packages to be used will be imported here.braries Needed Throught out 

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
````

## Step 2: Load the data

We have three datasets to work on 
1. Linearly Seperable dataset
<img src="images/LS.png" style="margin:0px">

2. Spiral dataset
<img src="images/interlock.png" style="margin:0px">

3. and, Interlock dataset.
<img src="images/spiral.png" style="margin:0px">

The datasets are two dimensional in nature and have only two classes (3 classes in case of LS). We have chosen these so that the concepts become clear and also for the reason that we can plot them.

The data points of the two classes are stored in two separate files. We will load the data points into a numpy array and label them artificially as `0` for class 1 and `1` for the other class.

The datasets are located in the `data` folder. Please check the data folder.

We ask User for input and depends on that input we choices the Dataset to Work on.

````python
d=input('Enter the data to be used: 1)LS 2)Interlock 3)Spiral \n')
d = int(d)

if d==1:
    fpath="data/LS_Group2/" 
elif d==2:
    fpath="data/interlock/"
else:
    fpath="data/spiral/" 
    
X_c1 = np.loadtxt(fpath+"Class1.txt")
X_c2= np.loadtxt(fpath+"Class2.txt")
y_c1=np.ones((X_c1.shape[0],1))
y_c2=np.zeros((X_c2.shape[0],1))
````

## Step 3: Split the data

We split the data into two splits: `train` - 75% and `test`  25%. Recall the import above 
```
from sklearn.model_selection import train_test_split
```
We will use this to split the dataset. Please google for its prototype and options.

But We stil Dont have both classes in same array we will use
```
np.concatenate
```
To Concatenate Both class in one Array


````python
X_c1_train, X_c1_test, y_c1_train, y_c1_test = train_test_split(X_c1, y_c1, test_size = 0.25, random_state = 0)
X_c2_train, X_c2_test, y_c2_train, y_c2_test = train_test_split(X_c2, y_c2, test_size = 0.25, random_state = 0)

X_train=np.concatenate((X_c1_train, X_c2_train), axis=0)
y_train=np.concatenate((y_c1_train, y_c2_train), axis=0)

X_test=np.concatenate((X_c1_test, X_c2_test), axis=0)
y_test=np.concatenate((y_c1_test, y_c2_test), axis=0)
````

## Step 4: Visualize the data

We are going to use `matplotlib.pyplot` which provides a MATLAB like 2d plotting interface.

````python
plt.figure()
plt.scatter(X_train[:,0],X_train[:,1],c=y_train.ravel(),cmap='rainbow')
plt.title("Data points of 2 classes")
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
````

## Step 5: Create the ML model object KNeighborsClassifier

In this case we will create a KNN object, however, this process can be repeated for any classifier. An important parameter here is the value of `K`. We take from the user.

````python
K=int(input('Enter the value of K \n'))

model = KNeighborsClassifier(n_neighbors = K, metric = 'minkowski', p = 2)
#The default metric is minkowski, and with p=2 is equivalent to the standard Euclidean metric.
````

## Step 6: Fitting the data by calling the `fit()` function on the training data

````python
model.fit(X_train, y_train.ravel())
````

## Step 7: Evaluate the model on the Test data.

We use `Predict()` Method to evaluate How good Over Model is on Test data 
If results are not good youcan go to setp 5 and 6 and make some changes in parameters etc.

````python
y_pred = model.predict(X_test)
````

### Printing The Confusion Matrix and Accuray 

#### 1.Confusion Matrix

<img src="images/CM.png" style="height:250px ; margin:0px" >
It is a `2 x 2` Matrix , <br>

   1. in `upper left` it contain `True Positive` <br>
       A true positive is an outcome where the model correctly predicts the positive class.<br>
       
   2. in `upper Right` it contain `False Positive`<br>
       A false positive is an outcome where the model incorrectly predicts the positive class.<br>

   3. in `Lower Rigth` it contain `True Negative` <br>
        true negative is an outcome where the model correctly predicts the negative class.<br>

   4. in `Lower Left` it contain `True Negative` <br>
        false negative is an outcome where the model incorrectly predicts the negative class.<br>
        
#### 2. Accuracy
Simply is the Percentage of How accurate over model is `i.e` how many it got `correct/incorrect*100`

````python
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
print("Confusion Matrix = \n", cm)
plt.imshow(cm, cmap=plt.cm.Blues)
print("The Accuracy=",accuracy*100)

````

##  Decision Boundaries

Here we will first create an artificial set of points covering the whole plot region and predict their labels using the trained model. Then the predicted label will be used to colour that point. The process will give us the decision surfaces. 

Once we get the decision surfaces, we will plot the test data to see if we have any mismatches.
<div class="alert alert-block alert-warning">
This will take some time to plot. Please adjust the step value for quicker plotting
</div>

### Visualising the Training set results

````python
cmap_1 = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])

X_set, y_set = X_train, y_train
plt.figure(1)
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))

Z=model.predict(np.c_[X1.ravel(), X2.ravel()])
Z = Z.reshape(X1.shape)
plt.pcolormesh(X1, X2, Z,cmap=cmap_1)
plt.scatter(X_train[:,0],X_train[:,1],c=y_train.ravel(),cmap='rainbow') 
plt.title('Decision region using KNN with K='+str(K))
plt.xlabel('X1')
plt.ylabel('X2')
````

### Visualising the Test set results

````python
plt.figure(2)
plt.pcolormesh(X1, X2, Z,cmap=cmap_1)
plt.scatter(X_test[:,0],X_test[:,1],c=y_test.ravel(),cmap='rainbow') 
plt.title('KNN (Test set)')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
````
