```python
def create_parameters():
  
  ###CREATING PARAMETER DICT
  
  
  #layer Size
  layers = [4 , 3] #Conv , #Dense
  
  #Filters
  filters = [32,32,32,64]
  
  #filter Size
  kernel_size = [ 4  , 4 , 4 , 4]
  
  #padding [ ['Conv' , 'MaxPool'] ]
  padding = [ ['same' , 'same'] , ['same' , 'same'] , ['same' , 'same'] , ['same' , 'same'] ]
  
  #pool_size
  
  pool_size = [2,2,2,2]

  #Nodes in Dense Layers
  dense_layers_size = [2048 , 100 , 5]

  #Activation for all Layers
  activation = ['relu' , 'relu' ,'relu' ,'relu' ,'relu' , 'relu','softmax' ]
  
#   dropout = [0.2,0.2 , 0.2 , 0.2]
#   dropout = [0 , 0 , 0 ,0]
  
  parameterDict = {
      'layers' : layers,
      'filters' : filters,
      'kernel_size' : kernel_size,
      'padding': padding,
      'pool_size':pool_size,
      'dense_layers_size' : dense_layers_size,
      'activation' : activation,
      'dropout' : False
#       'flsttenDrop' : False
  }
  
  
  ###CREATING CALLBACK DICT
  
  callbackDict = {
      'custom' : False,
      'tensorboard' : False,
      'modelcheckpoint' : True,
      'modelcheckpoint_path' : '/content/gdrive/My Drive/SeedApp/data/SavedModels/'
  }
  print("{} Conv Layers Got Created and {} Dense Layers".format(layers[0] , layers[1]))
  if len(filters) != layers[0]:
    print("Layer Number Dont Match {} filter should be equal to {} layers".format(len(filters) , layers[0]) )
  if len(padding) != layers[0]:
    print("Layer Number Dont Match {} Padding should be equal to {} layers".format(len(padding) , layers[0]) )
  if len(activation) != layers[0] + layers[1]:
    print("Activation dont match number of Layers {} activation != {} layers".format(len(activation) , layers[0] + layers[1] ) )
    
  return parameterDict , callbackDict
  
parametersDict1 , callbackDict1 = create_parameters()

```

```python
def create_parameters():
  
  ###CREATING PARAMETER DICT
  
  
  #layer Size
  layers = [6 , 3] #Conv , #Dense
  
  #Filters
  filters = [32,32,32,64,64,64]
  
  #filter Size
  kernel_size = [4 , 4 , 4  , 4 , 4 , 4]
  
  #padding [ ['Conv' , 'MaxPool'] ]
  padding = [ ['same' , 'same'] , ['same' , 'same'] , ['same' , 'same'] , ['same' , 'same'] , ['same' , 'same'] , ['same' , 'same'] ]
  
  #pool_size
  
  pool_size = [2,2,2,2,2,2]

  #Nodes in Dense Layers
  dense_layers_size = [512 , 50 , 5]

  #Activation for all Layers
  activation = ['relu' ,'relu' , 'relu' , 'relu' ,'relu' ,'relu' ,'relu' , 'relu','softmax' ]
  
#   dropout = [0.2,0.2 , 0.2 , 0.2]
#   dropout = [0 , 0 , 0 ,0]
  
  parameterDict = {
      'layers' : layers,
      'filters' : filters,
      'kernel_size' : kernel_size,
      'padding': padding,
      'pool_size':pool_size,
      'dense_layers_size' : dense_layers_size,
      'activation' : activation,
      'dropout' : False
  }
  
  
  ###CREATING CALLBACK DICT
  
  callbackDict = {
      'custom' : False,
      'tensorboard' : False,
      'modelcheckpoint' : True,
      'modelcheckpoint_path' : '/content/gdrive/My Drive/SeedApp/data/SavedModels/'
  }
  print("{} Conv Layers Got Created and {} Dense Layers".format(layers[0] , layers[1]))
  if len(filters) != layers[0]:
    print("Layer Number Dont Match {} filter should be equal to {} layers".format(len(filters) , layers[0]) )
  if len(padding) != layers[0]:
    print("Layer Number Dont Match {} Padding should be equal to {} layers".format(len(padding) , layers[0]) )
  if len(activation) != layers[0] + layers[1]:
    print("Activation dont match number of Layers {} activation != {} layers".format(len(activation) , layers[0] + layers[1] ) )
    
  return parameterDict , callbackDict
  
parametersDict3 , callbackDict3 = create_parameters()

```
