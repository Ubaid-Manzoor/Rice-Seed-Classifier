```
import random
import shutil
import os
    
def split_data(main_folder_name=['training' , 'test' , 'validation'] 
              ,sub_folder_name = ['C1' , 'C2' , 'C3' , 'C4' , 'C5'] 
              ,current_folder_path = '/content/gdrive/My Drive/SeedApp/data/input' 
              ,new_folder_path = '/content/gdrive/My Drive/SeedApp/data/input' 
              ,split_ratio = [0.7 , 0.15]
             ):
  random.seed(0)
  
  train_len = 0
  test_len = 0
  validation_len = 0
  #Store Full Path of Folder to be Created
  training = os.path.join(new_folder_path , main_folder_name[0]) 
  test = os.path.join(new_folder_path , main_folder_name[1]) 
  validation = os.path.join(new_folder_path , main_folder_name[2]) 
    
    #Create New Folders
  for main_folder in main_folder_name:

    current_main_folder = os.path.join(new_folder_path , main_folder)

    if not os.path.exists(current_main_folder):
      os.mkdir(current_main_folder)
    for sub_folder in sub_folder_name:
      if not os.path.exists(os.path.join(current_main_folder , sub_folder)):
        os.mkdir(os.path.join(current_main_folder , sub_folder))

  #Fill the Folder with Img data
  for sub_folder in sub_folder_name:
    current_class_folder = os.path.join(current_folder_path , sub_folder)

    img_file_name = os.listdir(current_class_folder)
    random.shuffle(img_file_name)



    training_img_len = int( ((len(img_file_name))) * split_ratio[0] )
    test_img_len = int( ((len(img_file_name))) * split_ratio[1] )
    validation_img_len = int( (len(img_file_name) -  training_img_len - test_img_len) )
    
    train_len += training_img_len
    test_len += test_img_len
    validation_len += validation_img_len
    


    training_img_list =  img_file_name[0:training_img_len]
    test_img_list =  img_file_name[training_img_len:training_img_len+test_img_len]
    validation_img_list =  img_file_name[training_img_len+test_img_len:-1]

    img_list = [training_img_list , test_img_list , validation_img_list]

    for i in range(len(img_list)):
      current_list = img_list[i]
      current_main_dir = main_folder_name[i]
      for img_path in current_list:
        source = os.path.join(current_folder_path , sub_folder , img_path)
        dest = os.path.join(new_folder_path , current_main_dir ,sub_folder )
        if(not os.path.isfile(os.path.join(dest , img_path) ) ):
          shutil.copy(source , dest)
        else:
          print(os.path.join(dest , img_path) , " :  Exists")
          
  
  return train_len , test_len , validation_len
```
