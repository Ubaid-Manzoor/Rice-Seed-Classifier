```
#############Apply Transform#################
import cv2
import os
print("End")

class transformImages():
  
  # theta
  # tx
  # ty
  # shear
  # zx
  # zy
  # flip_horizontal
  # flip_vertical
  # channel_shift_intencity
  # brightness
  
  
  @classmethod
  def saveImagesByClass(cls,X , save_to_dir , prefix):
    for index,img in enumerate(X):
      cv2.imwrite(os.path.join(save_to_dir , prefix+str(index) + '.png' ) , cv2.cvtColor(img , cv2.COLOR_BGR2RGB))


  @classmethod 
  def saveImages(cls , Images ,Y, main_dir , prefix  , classes):
    for i in range(classes):
        ith_class_images = Images[Y == i]

        print(ith_class_images.shape)
          
        save_to_dir = os.path.join(main_dir , 'C' + str(i+1) )

        if (not os.path.exists(save_to_dir) ):
          os.mkdir(save_to_dir)
        
        cls.saveImagesByClass(ith_class_images , save_to_dir , prefix)
        
        
  ####################     ROTATE     #################################
  @classmethod
  def rotate(cls ,X ,Y=None, save_to_dir = None , prefix = 'rotatedImage' , classes=5):
    
    RotatedImages = []
    
    image_gen = ImageDataGenerator()
    for image in X:
      RotatedImage = image_gen.apply_transform(image, transform_parameters = {'theta' : random.randrange(30,90)})
      RotatedImages.append(RotatedImage)
    
    RotatedImages = np.array(RotatedImages)
    
    if save_to_dir != None:
      save_to_dir = os.path.join(save_to_dir , 'RotatedImages')
      if(not os.path.exists(save_to_dir)):
        os.mkdir(save_to_dir)
      main_dir = save_to_dir
    
      cls.saveImages(RotatedImages , Y , main_dir , prefix , classes)
     
    return RotatedImages
  
  ####################     SHIFT     #################################
  @classmethod
  def shift(cls ,X ,Y=None, save_to_dir = None , prefix = 'ShiftedImage' , classes=5):
    
    ShiftedImages = []

    image_gen = ImageDataGenerator()
    for image in X:
      choise = random.randrange(0,2)
      if choise == 1:
        ShiftedImage = image_gen.apply_transform(image, transform_parameters = {'tx' : random.randrange(15 , 55)  , 'ty' : random.randrange(10,25)*-1 })
      else:
        ShiftedImage = image_gen.apply_transform(image, transform_parameters = {'tx' : random.randrange(15 , 55)  , 'ty' : random.randrange(10,25) })
      ShiftedImages.append(ShiftedImage)
    
    ShiftedImages = np.array(ShiftedImages)
    
    if save_to_dir != None:
      save_to_dir = os.path.join(save_to_dir , 'ShiftedImages')
      if(not os.path.exists(save_to_dir)):
        os.mkdir(save_to_dir)
      main_dir = save_to_dir
      cls.saveImages(ShiftedImages , Y , main_dir , prefix , classes)
     
    return ShiftedImages
  
  ####################     ZOOM     #################################
  @classmethod
  def zoom(cls ,X ,Y=None, save_to_dir = None , prefix = 'ZoomedImage' , classes=5):
    ZoomedImages = []

    image_gen = ImageDataGenerator()
    for image in X:
      value_to_zoom = random.randrange(10,40) / 10
      ZoomedImage = image_gen.apply_transform(image, transform_parameters = {'zx' : value_to_zoom  , 'zy' : value_to_zoom})
      ZoomedImages.append(ZoomedImage)
    
    ZoomedImages = np.array(ZoomedImages)
    
    if save_to_dir != None:
      save_to_dir = os.path.join(save_to_dir , 'ZoomedImages')
      if(not os.path.exists(save_to_dir)):
        os.mkdir(save_to_dir)
      main_dir = save_to_dir
      cls.saveImages(ZoomedImages , Y , main_dir , prefix , classes)
     
    return ZoomedImages
  
  ####################     FLIP     #################################
  
  @classmethod
  def flip(cls ,X ,Y=None, save_to_dir = None , prefix = 'FilpedImage' , classes=5):
    FlipedImages = []

    image_gen = ImageDataGenerator()
    for image in X:
      FlipedImage = image_gen.apply_transform(image, transform_parameters = {'flip_horizontal' : True ,'flip_vertical' : True})
      FlipedImages.append(FlipedImage)
    
    FlipedImages = np.array(FlipedImages)
    
    if save_to_dir != None:
      save_to_dir = os.path.join(save_to_dir , 'FlipedImages')
      if(not os.path.exists(save_to_dir)):
        os.mkdir(save_to_dir)
      main_dir = save_to_dir
      
      cls.saveImages(FlipedImages , Y , main_dir , prefix , classes)
     
    return FlipedImages
  
  ####################     BRIGHTNESS     #################################
  
  @classmethod
  def changeBrightness(cls ,X ,Y=None, save_to_dir = None , prefix = 'BrightedImage' , classes=5):
    BrightedImages = []

    image_gen = ImageDataGenerator()
    for image in X:
      BrightedImage = image_gen.apply_transform(image, transform_parameters = {'brightness' :  random.randrange(6,14) / 10})
      BrightedImages.append(BrightedImage)
      
    BrightedImages = np.array(BrightedImages)
    BrightedImages = BrightedImages.astype(int) 
    
    if save_to_dir != None:
      save_to_dir = os.path.join(save_to_dir , 'BrightedImages')
      if(not os.path.exists(save_to_dir)):
        os.mkdir(save_to_dir)
      main_dir = save_to_dir
      cls.saveImages(BrightedImages , Y , main_dir , prefix , classes)
     
    return BrightedImages
  
  
  
```
